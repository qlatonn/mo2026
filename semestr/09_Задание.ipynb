## Задание 1.

Прочитать данные
`https://raw.githubusercontent.com/yakushinav/omo/main/data/movies.csv`

Имена столбцов
```
['color', 'director_name', 'num_critic_for_reviews', 'duration',
                'gross', 'movie_title', 'num_user_for_reviews', 'country',
```

````python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/yakushinav/omo/main/data/movies.csv',
                sep='|',
                header=None,
                na_values=['','?','-'],
                names= ['color', 'director_name', 'num_critic_for_reviews', 'duration',
                'gross', 'movie_title', 'num_user_for_reviews', 'country',
                'content_rating', 'budget', 'title_year', 'imdb_score', 'genre'],
                thousands=',')

df.head()
````

## Задание 3

Прочитать данные `https://raw.githubusercontent.com/yakushinav/omo/main/data/artists.json`.

Выполнить сохранение в формат CSV. Прочитать полученный файл. Сравнить с исходным.


````python
url_artists = "https://raw.githubusercontent.com/yakushinav/omo/main/data/artists.json"

df_artists = pd.read_json(url_artists)

print("Исходные данные (JSON):")
print(df_artists.head())

csv_filename = "artists.csv"
df_artists.to_csv(csv_filename, index=False)

df_artists_from_csv = pd.read_csv(csv_filename)

print("\nДанные из CSV:")
print(df_artists_from_csv.head())

are_equal = df_artists.equals(df_artists_from_csv)
print(f"\nDataFrame равны? {are_equal}")

print("\nТипы данных в исходном JSON:")
print(df_artists.dtypes)
print("\nТипы данных в прочитанном CSV:")
print(df_artists_from_csv.dtypes)
````
